```python
import numpy as np 
import matplotlib.pyplot as plt
from kafe2 import  XYContainer, Fit, XYFit, Plot, ContoursProfiler  
import pandas as pd
```

```python
data=pd.read_csv('Aufgabe 3.txt',delimiter=';',decimal=".",skiprows=1) #importiere Daten
data.columns=['w','wN']
w = data['w'].to_numpy() 
wN = data['wN'].to_numpy()
w = w * 2 * np.pi
wN = wN *2 * np.pi

xy_data = XYContainer(w,wN)
line_fit=Fit(data=xy_data)
line_fit.do_fit()
line_fit.assign_parameter_names(x='w in 1/s', a='a', b='b')
line_fit.assign_model_function_expression('wN')
line_fit.assign_model_function_expression("{a} * {x} + {b}")
line_fit.report()

line_fit.assign_parameter_latex_names(x='w', a='a', b='b')
line_fit.assign_model_function_latex_name('wN')
line_fit.assign_model_function_latex_expression('{a} \\cdot {x} + {b}')


line_fit.data_container.label = "Verwendete Daten"
line_fit.model_label = "Modellfunktion"

line_fit.data_container.axis_labels = ["w in 1/s", "wN in 1/s"]

plot = Plot(fit_objects=line_fit)
plot.plot()
plot.show()
plot.save('Fitw1.png')
```

```python
data=pd.read_csv('Aufgabe 4.txt',delimiter=';',decimal=".",skiprows=1) #importiere Daten
data.columns=['t','f']
t = data['t'].to_numpy() 
f = data['f'].to_numpy()

plt.plot(t,f,'.')
plt.xlabel('t in s')
plt.ylabel('w in 1/s')
plt.savefig('frequenzdaempfung.png')

def expfit(x, a=30,k=0,b=0):
    return a*np.exp(-k*x)-b*x
t = t[:]
f = f[:]
w = f * 2 * np.pi
xy_data = XYContainer(t,w)
line_fit=Fit(data=xy_data,model_function=expfit)
line_fit.do_fit()
line_fit.assign_parameter_names(x='w in 1/s', a='a', b='b')
line_fit.assign_model_function_expression('w')
#line_fit.assign_model_function_expression("{a} * {x} + {b}")
#line_fit.report()

line_fit.assign_parameter_latex_names(x='t', a='a', b='b')
line_fit.assign_model_function_latex_name('w')
#line_fit.assign_model_function_latex_expression('{a} \\cdot {x} + {b}')


line_fit.data_container.label = "Verwendete Daten"
line_fit.model_label = "Modellfunktion"

line_fit.data_container.axis_labels = ["t in s", "w in 1/s"]

plot = Plot(fit_objects=line_fit)
plot.plot()
plot.show()
plot.save('Fit4mitdaempfung.png')
```

```python
data=pd.read_csv('Aufgabe 5.txt',delimiter=',',decimal=".",skiprows=1) #importiere Daten
data.columns=['f1','t','f2','fm']
f1 = data['f1'].to_numpy() 
t = data['t'].to_numpy()
f2 = data['f2'].to_numpy()
fm = data['fm'].to_numpy()
wp = (2*np.pi)/t
invwm = 1/(2*np.pi*fm)
plt.plot(fm,t)
plt.xlabel('Mittlere Frequenz in 1/s')
plt.ylabel('Przzesionsdauer in s')
plt.savefig('Prazzesionsdauer.png')
xy_data = XYContainer(invwm,wp)
line_fit=Fit(data=xy_data)
line_fit.do_fit()
line_fit.assign_parameter_names(x='1/Mittlere Kreisrequenz in s', a='a', b='b')
line_fit.assign_model_function_expression('w_p')
line_fit.assign_model_function_expression("{a} * {x} + {b}")
#line_fit.report()

line_fit.assign_parameter_latex_names(x='1/w_m', a='a', b='b')
line_fit.assign_model_function_latex_name('w_p')
line_fit.assign_model_function_latex_expression('{a} \\cdot {x} + {b}')


line_fit.data_container.label = "Verwendete Daten"
line_fit.model_label = "Modellfunktion"

line_fit.data_container.axis_labels = ["1/Mittlere Kreisfrequenz in s", "w_p in 1/s"]

plot = Plot(fit_objects=line_fit)
plot.plot()
plot.show()
plot.save('Fit5.png')

```
