```python
xy_data = XYContainer(t6_5v400mA,s6_5v400mA)
def cosfit(t, x=1, w=3.306939635357677, phi=10):
    return w * np.sin(x*t + phi) 
line_fit=Fit(data=xy_data,model_function=cosfit)
#line_fit.add_error('y',0.05)
#line_fit.add_error('x',0.05)
line_fit.do_fit()
line_fit.data_container.label = "Verwendete Daten"
line_fit.model_label = "Modellfunktion"

plot = Plot(fit_objects=line_fit)
plot.plot()
```
