## Aufgabe 3: Ein einfaches Modell zur Bestimmung von $g$

Wir legen unserer ersten Messung das einfache Modell eines [mathematischen Pendels](https://de.wikipedia.org/wiki/Mathematisches_Pendel) zugrunde. 
$$
\begin{split}
& m\,l^{2}\,\,\ddot{\phi} + m\,g\,l\,\phi = 0, \\
\end{split}
$$
wobei $m$ einer Punktmasse und $l$ dem Abstand der Punktmasse vom Angelpunkt des Pendels entsprechen. Diese Differentialgleichung wird von harmonischen Schwingungen der Form 
$$
\begin{split}
& \phi(t) = \phi_{0}\cos\left(\omega\,t + \delta\right) \\
\end{split}
$$
gelöst, wobei $\phi_{0}$ die Amplitude, $\omega$ die Kreisfrequenz und $\delta$ eine freie Phase der Schwingung sind. Mit Hilfe dieses Modells können Sie bei gegebenem $l$ die Größe von $g$ mit Hilfe der Gleichung
$$
\begin{equation}
g = \frac{4\,\pi^{2}}{T^{2}}l
\end{equation}
$$
aus der Periode $T$ der Schwingung bestimmen. Zusätzliche Information, die Sie zur Lösung dieser Aufgabe benötigen, finden Sie in der Datei

## Lösung:
### Aufgabe 3.1:

Die zu erwartende Erdbeschleunigung hat einen Literaturwert von $g_{exp}=9.809599 \pm 0.000034 \frac{m}{s^2}$

````python
autocov = np.correlate(x, x, mode='full' )[-len(x):] #Autokorrelation berechnen
autocorr = autocov / autocov[0]
````

````python
peaks=ppk.convolutionPeakfinder(autocorr)            #Peaks herausfinden
peaks = peaks[1:-1]
````

````python
fig2 = plt.figure(2,figsize=(10,5))                 #Plotten
ax2 = fig2.add_subplot(2,1,1)
ax2.plot(t,autocorr)
ax2.set_xlabel("Zeit in s")
ax2.set_ylabel("Amplitude")

#Periodendauer bestimmen
t_mess=np.zeros(len(peaks)-1)
i=0
currenttime=0
for peak in peaks:
    if(i>0&i<len(peaks)-1):
        plt.vlines(t[peak], -1, 1, color='red', ls='--')
        t_mess[i-1]=t[peak].copy()-currenttime
    i=i+1
    currenttime=t[peak].copy()


print('Mittelwert Periodendauer:',np.mean(t_mess))     #Mittelwert Periodendauer
print('Standardabweichung der Periodendauer',np.sqrt(np.var(t_mess)))#Standardabweichung 
````

````python
T=ufloat(np.mean(t_mess),np.sqrt(np.var(t_mess))) #Periodendauer speichern
l=ufloat(0.6285,0.6285-0.6280)                    #Pendellänge speichern
````

````python
g=((4*(np.pi)**2)/T**2)*l                       #g berechnen
print('Die Erdbeschleunigung hat den Wert:',g,'m/s^2')
````

Daraus ermittlen wir: 
Mittelwert der Erdbeschleunigung $g$: 9.701 $\frac{m}{s^2}$ $\\$
Unsicherheit der Erdbeschleunigung $\Delta g$: 0.008 $\frac{m}{s^2}$

Weder der Mittelwert noch die Unsicherheit der Erdbeschleunigung passt mit dem Literaturwert überein.

````python
yfit = afit*unumpy.cos(omegafit*t+phifit)              #bei verbesserter Funktion Peaks berechnen  
peaksmax= np.array(ppk.convolutionPeakfinder(yfit)[:-1])
peaksmin = np.array(ppk.convolutionPeakfinder(-yfit)[:-1])
print('{} Extrema gefunden!'.format(len(peaksmin)+len(peaksmax)))
plt.plot(t,autocorr)
plt.xlabel('Zeit(s)')
plt.ylabel('Autokorrelation')
plt.show()
#array aus zeitdifferenz der peaks bilden; die unsicherheit davon ist die unsicherheit der Frequenz


peakdiff=np.array([],dtype=float)
peaksmaxdiff = peaksmax[1:] - peaksmax[0:-1]
peaksmindiff = peaksmin[1:] - peaksmin[0:-1]
#print(peaksmaxdiff)
#print(peaksmindiff)

peakdiff = np.concatenate((peaksmaxdiff,peaksmindiff),dtype=float) #arrays zusammenführen


peakdiff = peakdiff / abtastrate

Tmean = np.mean(peakdiff)
Tstd = np.sqrt(np.var(peakdiff,ddof=1))
print(f"Mittelwert von T: {Tmean} , Standartabweichung T: {Tstd}")
Tb = ufloat(Tmean,Tstd)

g1=((4*(np.pi)**2)/Tb**2)*l
print('g=',g1,'m/s^2')
````

### Aufgabe 4.3:

Modellfunktion mit Dämpfung:

  = \left(\omega^2+\frac{1}{\tau}\right)\frac{I}{m\,\ell}
